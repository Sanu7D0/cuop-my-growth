```python
import pandas as pd
import numpy as np
from IPython.display import display
```

### 인덱스, 칼럼 정리

```python
# 칼럼0 인덱스로, 로우0,1 칼럼명으로 불러오기
df = pd.read_csv('./dataset/stat_100701 - Sheet0.csv', encoding='utf-8', index_col=0, header=[0,1])
# df = pd.read_csv('./dataset/stat_100701 - Sheet0.csv', encoding='utf-8', index_col=0, header=[0,1], thousands=",")
display(df.head())

# 인구밀도 칼럼 드랍, 칼럼명(인구, 인구, 인구, ...) 드랍
df = df.drop("인구밀도", level=1, axis=1).droplevel(level=1, axis=1)
display(df.head())

# 합계, 수도권 로우 드랍
df.drop(["계", "수도권"], inplace=True)
display(df.head())
```

### 전처리

```python
# 결측치 "-" -> 0
df.replace("-", 0, inplace=True)

# 쉼표 제거, 정수화
df = df.applymap(lambda x: int(str(x).replace(",", "")))
df.head()
```

### 지역별 합산 구하기

```python
# 지역구분 칼럼을 추가
df.insert(loc=0, column="지역구분", value=["비수도권"] * len(df))

# 수도권은 수도권으로 표시
captial_area = ["서울", "경기", "인천"]
for area in captial_area:
    df.loc[area, "지역구분"] = "수도권"
df.head()
```

```python
df_sum = df.groupby(by=["지역구분"]).sum()
# df_sum.applymap(lambda x: x * 1000) # 1000명 단위
display(df_sum)

# 완성!
df_result = df_sum.T.reset_index()
df_result.columns = ["연도", "비수도권 인구", "수도권 인구"]
df_result
```

# 더 쉬운 방법

```python
df = pd.read_csv('./dataset/stat_100701 - Sheet0.csv', encoding='utf-8', index_col=0, header=[0,1])
df = df.drop("인구밀도", level=1, axis=1).droplevel(level=1, axis=1)
df.drop(["계", "수도권"], inplace=True)

# 전처리
df.replace("-", 0, inplace=True)
df = df.applymap(lambda x: int(str(x).replace(",", "")))

# 뽑아오기
captial_area = ["서울", "경기", "인천"]
capital_sum =  df.loc[captial_area].sum()
non_capital_sum = df.loc[list(set(list(df.index)) - set(captial_area))].sum()

df_result =  pd.concat([capital_sum, non_capital_sum], axis=1)
df_result.columns = ["수도권 인구", "비수도권 인구"]
df_result
```
