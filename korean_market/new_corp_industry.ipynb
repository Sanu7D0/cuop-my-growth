```python
import pandas as pd
import numpy as np
from IPython.display import display

import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
df = pd.read_csv("./dataset/업종별_창업기업수_20220715110806.csv", encoding="cp949", header=[0,1], index_col=[0,1])
df.drop(["법인", "개인"], level=1, axis=1, inplace=True)
df = df.droplevel(level=1, axis=1)#.drop("합계", level=0, axis=0)
df = df.query("ilevel_1 == '소계'")
df = df.T.droplevel(level=1, axis=1)
df.index.name = None

fig = go.Figure()
for col in df.columns:
    fig.add_trace(go.Scatter(x=df.index, y=df[col], mode="lines", name=col))
fig.update_layout(title="업종별 창업 기업 수")
fig.show()
```

```python
df = pd.read_csv("./dataset/업종별_창업기업수_20220715110806_소계제거.csv", encoding="cp949", header=[0,1], index_col=[0,1])
df = df.drop(["법인", "개인"], level=1, axis=1).droplevel(level=1, axis=1).drop("합계", level=0, axis=0)
df = df.T
df.index.name = None

fig = go.Figure()
for col in df.columns:
    fig.add_trace(go.Scatter(x=df.index, y=df[col[0]][col[1]], mode="lines", name=col[1]))
fig.show()
```
